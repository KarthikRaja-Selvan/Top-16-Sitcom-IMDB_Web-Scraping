```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#The Simpsons
Season = []
Episode = []
Title = []
Description = []
Episode_AirDate = []
Episode_Rating = []
Total_Voters = []

for season in range(33):
    
    season_no = season + 1
    url = 'https://www.imdb.com/title/tt0096697/episodes?season=%s' % season_no

    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')

    lists = soup.find_all('div', {'class': 'info'})
    
    for episode_no, list in enumerate(lists, start=1):
        Season_index = str(season_no)
        title = list.find('a', {'itemprop': 'name'}).text.strip()
        description = list.find('div',{'class':'item_description'}).text.strip()
        air_date = list.find('div', {'class': 'airdate'}).text.strip()
        avg_rating = list.find('span', {'class': 'ipl-rating-star__rating'}).text.strip()
        total_rating = list.find('span', {'class': 'ipl-rating-star__total-votes'}).text.strip()
        
        Season.append(Season_index)
        Episode.append(episode_no)
        Description.append(description)
        Title.append(title)
        Episode_AirDate.append(air_date)
        Episode_Rating.append(avg_rating)
        Total_Voters.append(total_rating)
        
    
    info = {'Season':Season,'Episode':Episode,'Title':Title,'Description':Description,'AirDate':Episode_AirDate,'Rating':Episode_Rating,'Voters':Total_Voters}
        
    df = pd.DataFrame(info)

df.to_csv('the_simpsons.csv', index=False)
```
